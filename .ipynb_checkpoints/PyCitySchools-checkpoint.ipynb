# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school. 
---

```python
# dependencies and setup
import pandas as pd
import numpy as np
import os

# files to Load
school_data_filepath = os.path.join("Resources", "schools_complete.csv")
student_data_filepath = os.path.join("Resources", "students_complete.csv")

# store School and Student Data files into Pandas DataFrames
school_data = pd.read_csv(school_data_filepath)
student_data = pd.read_csv(student_data_filepath)

# combine the data into a single dataset
district_data_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
district_data_df.head()
# district_data_df.count()
# district_data_df.dtypes
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Hold the calculations in a DataFrame

```python
total_schools = district_data_df['school_name'].nunique()
# total_schools

total_students = district_data_df['student_name'].count()
# total_students

#school_data.head()
district_budget = school_data['budget'].sum()
# district_budget

average_math_score = district_data_df['math_score'].mean()

average_reading_score = district_data_df['reading_score'].mean()

passing_rate = (average_math_score + average_reading_score)/2



students_passing_math = 0
for i in district_data_df['math_score']:
    if i >=70:
        students_passing_math = students_passing_math + 1

pct_passing_math = (students_passing_math/int(total_students))*100
# pct_passing_math 


students_passing_reading = 0
for i in district_data_df['reading_score']:
    if i >=70:
        students_passing_reading = students_passing_reading + 1
        
pct_passing_reading = (students_passing_reading/int(total_students))*100
# pct_passing_reading
```

```python
district_summary_dict = {'Total Schools': total_schools, 'Total Students': total_students, 'Total Budget': district_budget, 
                         'Average Math Score': average_math_score, 'Average Reading Score': average_reading_score, 
                         '% Passing Math': pct_passing_math, '% Passing Reading': pct_passing_reading, 
                         'Overall Passing Rate': passing_rate}

district_summary_df=pd.DataFrame([district_summary_dict])

district_summary_df = district_summary_df[['Total Schools', 'Total Students', 'Total Budget', 'Average Math Score',
                                          'Average Reading Score', '% Passing Math', '% Passing Reading', 
                                          'Overall Passing Rate']]

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# per_student_budget
school_data_new = school_data
school_data_new['per_student_budget'] = school_data_new['budget']/school_data_new['size']
school_data_df = school_data_new.set_index('school_name')
school_data_df.head()
```

```python
# get the average score for reading and math 
district_group = district_data_df.groupby(['school_name']).size()
district_group

district_group_avg = district_data_df.groupby(['school_name']).mean()
district_group_avg

avg_reading = district_group_avg['reading_score'].rename('avg reading score')
avg_math = district_group_avg['math_score'].rename('avg math score')
```

```python
# get # of studnets in each school who passed math
passing_math = district_data_df[district_data_df.math_score >69].groupby(['school_name']).size()
passing_math = passing_math.rename('passed math').reset_index()
# print(passing_math)
```

```python
# get # of studnets in each school who passed reading
passing_reading = district_data_df[district_data_df.reading_score >69].groupby(['school_name']).size()
passing_reading = passing_reading.rename('passed reading').reset_index()
# print(passing_reading)
```

```python
# get total studnets in each school
school_student_total = district_data_df.groupby(['school_name']).size()
school_student_total = school_student_total.rename('total_students').reset_index()
```

```python
#merge dataframes: number of students in each school who passed reading or math and the total number of students 
merged_passing = pd.merge(passing_reading, school_student_total, how="inner", on="school_name")
# merged_passing.head()

merged_passing = pd.merge(merged_passing, passing_math, how="inner", on=["school_name", "school_name"])
# merged_passing.head()

merged_passing = pd.merge(merged_passing, avg_reading, how="inner", on=["school_name", "school_name"])
merged_passing = pd.merge(merged_passing, avg_math, how="inner", on=["school_name", "school_name"])
merged_passing.head()
```

```python
# combine passing data for each school and get the %passed for each subject and the overall passing rate 
merged_passing['% Passing Reading'] = (merged_passing['passed reading']/merged_passing['total_students'])*100
merged_passing['% Passing Math'] = (merged_passing['passed math']/merged_passing['total_students'])*100
merged_passing['Overall Passing Rate'] = (merged_passing['% Passing Reading'] + merged_passing['% Passing Math'])/2
merged_passing.head()
```

```python
# merge passing data info with school only info to creat new dataframe
data = pd.merge(merged_passing, school_data_df, how='inner', on=['school_name'])
all_data = data[['school_name', 'type', 'total_students', 'avg reading score', 'avg math score', '% Passing Reading', '% Passing Math', 'Overall Passing Rate', 'per_student_budget', 'budget', 'School ID']]
all_data_df = all_data.set_index('school_name')
all_data_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
# all_data_df.sort_values(by=['per_student_budget'], inplace=True)
# all_data_df
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585+", "$585-615", "$615-645", "$645-675"]

all_data_df["Spending Group"] = pd.cut(all_data_df["per_student_budget"], spending_bins, labels=group_names)
scores_by_school_spending = all_data_df.groupby('Spending Group').mean()
scores_by_school_spending
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

all_data_df["school_size_range"] = pd.cut(all_data_df["total_students"], size_bins, labels=group_names)
all_data_df.groupby('school_size_range').mean()
```
